## **Для выбранного датасета (датасетов) на основе материалов лекции решить следующие задачи:**

- обработку пропусков в данных:

Проверяем формат данных

```python
print(data.dtypes)
```

Так же при проверке данных не было замечено каких либо признаков не корректности

```python
missing_values = data.isnull().sum()
count_values = data.count()
print(count_values)
print(missing_values)
```

Мы видим, что в столбце **total_bedrooms** имеется 207 пропущенных столбцов
Предлагаю вместо того, чтобы удалять пропущенные данные и не учитывать их, заменить их на среднее значение на основе не пропущенных данных


```python
data['total_bedrooms'] = data['total_bedrooms'].fillna(data['total_bedrooms'].mean())
missing_values = data.isnull().sum()
count_values = data.count()
print(count_values)
print(missing_values)
```

- кодирование категориальных признаков

```python
data = pd.get_dummies(data, columns=['ocean_proximity'], prefix='ocean_proximity')
```

```python
data.head(10)
```

```python
print(data.dtypes)
```

Мы разделили данные категориально на ocean, inland, island, near bay, near ocean с помощью one hot encoding. Тем самым теперь у категорий нет веса в отличии если бы мы делили с помощью ярлычного деления, которое присваивает значения 1 2 3 4, из-за этого компьютер думал бы что у категорий есть вес

- масштабирование данных


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled_df = pd.DataFrame(data_scaled, columns=data.columns)
print(data_scaled_df.head(10))
```
